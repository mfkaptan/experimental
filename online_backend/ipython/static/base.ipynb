```python
%%javascript

function get_modifiers(e) {
    var modifiers = [];
    if (e.altKey) modifiers.push("alt");
    if (e.ctrlKey) modifiers.push("ctrl");
    if (e.metaKey) modifiers.push("meta");
    if (e.shiftKey) modifiers.push("shift");
    return modifiers;
};

function gen_mouse_event(c, e, type) {
    var rect = c.getBoundingClientRect();
    var modifiers = get_modifiers(e);
    var event = {
        "source": "browser",
        "event":
        // Mouse Event
        {
            "name": "MouseEvent",
            "properties":
            {
                "type": type,
                "pos": [e.clientX-rect.left, e.clientY-rect.top],
                "button": e.button | null,
                "delta": null,
                "modifiers": modifiers,
            }
        }
    };
    return event;
};

function gen_key_event(c, e, type){
    var event = {
        "source": "browser",
        "event":
        // Key Event
        {
            "name": "KeyEvent",
            "properties":
            {
                "type": type,
                "key": e.which,
            }
        }
    };
    return event;
};

require(["widgets/js/widget"], function(WidgetManager){
    var Widget = IPython.DOMWidgetView.extend({
        render: function(){
            this.$canvas = $('<canvas />')
                .attr('id', 'canvas')
                .attr('width', '500')
                .attr('height', '500')
                .attr('tabindex', '1')
                .appendTo(this.$el);
        },
        
        events: {
            'mousemove': 'mouse_move',
            'mousedown': 'mouse_press',
            'mouseup': 'mouse_release',
            'mousewheel': 'mouse_wheel',
            'keydown': 'key_press',
            'keyup': 'key_release',
            
        },
        
        mouse_move: function(e) {
            var event = gen_mouse_event(this.$canvas[0], e, "mouse_move");
            this.send(event);
        },
        
        mouse_press: function(e) {
            var event = gen_mouse_event(this.$canvas[0], e, "mouse_press");
            this.send(event);
        },
        
        mouse_release: function(e) {
            var event = gen_mouse_event(this.$canvas[0], e, "mouse_release");
            this.send(event);
        },
        
        mouse_wheel: function(e) {
            var event = gen_mouse_event(this.$canvas[0], e, "mouse_wheel");
            event.events[0].properties.delta = [e.wheelDeltaX / 100., e.wheelDeltaY / 100.];
            this.send(event);
        },
            
        key_press: function(e) {
            var event = gen_key_event(this.$canvas[0], e, "key_press");
            this.send(event);
        },
            
        key_release: function(e) {
            var event = gen_key_event(this.$canvas[0], e, "key_release");
            this.send(event);
        },
        
        // Update, whenever value attribute of our widget changes
        update : function(){
            var canvas2d = this.$canvas[0].getContext('2d');
            var img = new Image();
            var istr = this.model.get("value");
            img.src = "data:image/png;base64," + istr;
            canvas2d.drawImage(img, 0, 0);
        },
    })
    WidgetManager.register_widget_view("Widget", Widget);
});
```
